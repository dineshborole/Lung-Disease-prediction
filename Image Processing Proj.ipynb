```python
import matplotlib.pyplot as plt
from skimage import io
```

```python
img1 = io.imread(r"xray_original.jpg")
plt.imshow(img1);
```

```python
img2 = io.imread(r"xray_modified.jpg")
plt.imshow(img2);
```

```python
img = img1[:,:,:]-img2[:,:,:]
plt.imshow(img);
```

```python
mask = img[:,:,:] > 248

img3 = img2.copy()
img3[mask[:,:,1],1] = 0 # Replacing all values in Green channel to 0 (Black)
img3[mask[:,:,1],2] = 0 # Replacing all values in Blue channel to 0 (Black)
plt.imshow(img3);
```
